```python
import pandas as pd
import json
```

```python
df = pd.read_csv('Datasets\survery_1.csv', sep=',')
df.head()
```

```python
df.rename(columns = {'Job title':'Job_title'}, inplace = True)
```

```python
df.rename(columns = {'What is your annual salary?':'Salary'}, inplace = True)
```

```python
df.rename(columns = {'Where are you located? (City/state/country)':'Location'}, inplace = True)
```

```python
df.rename(columns = {'Please indicate the currency':'currency'}, inplace = True)
```

```python
df.rename(columns = {'How many years of post-college professional work experience do you have?':'YearsExperience'}, inplace = True)
```

```python
df.head(50)
```

```python
data = df[["Job_title", "Salary","currency","Location","YearsExperience"]]
```

```python
data.head(60)
```

```python
data['Salary'] = data['Salary'].str.replace(',', '').str.replace('$', '').str.replace(' ', '').str.replace('.', '')
```

```python
data.head(50)
```

```python
data["Salary"].head(60)
```

```python
data['Salary']  = pd.to_numeric(data['Salary'] )
```

```python
with open('data.txt', 'w') as f:
    for row in  data.to_dict('records'):
        f.write(json.dumps(row) + '\n')
```
